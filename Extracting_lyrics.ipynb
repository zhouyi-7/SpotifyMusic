```python
token = credentials["lyricsgenius_token"]
```

```python
from lyricsgenius import Genius
genius = Genius(token)
artist = genius.search_artist("Andy Shauf", max_songs=3, sort="title")
```

```python
# Way 1
song = genius.search_song("To You", artist.name)

# Way 2
# this will search artist.songs first
# and if not found, uses search_song
song = artist.song("To You")

print(song.lyrics)
```

```python
%%time

lyrics = []

for i, r in df.iterrows():
    print(i)
    curr_song = r['Name']
    print(curr_song)
    curr_artist = r['Artist']
    print(curr_artist)
    try:
        song = genius.search_song(artist=curr_artist, title=curr_song)
        lyrics.append(song.lyrics)
    except Exception as e:
        s = "Error " + str(e)
        lyrics.append(s)
        
df['lyrics'] = lyrics       
```

```python
len([l for l in lyrics if l.startswith("Error")])
```

```python
df = df.drop('artist_clean', axis=1)
```

```python
df = df.drop('song_clean', axis=1)
```

```python
%%time

df['genius_lyrics'] = df.apply(lambda x: get_song_lyrics(genius, 
                                                         artist=x['Artist'], 
                                                         song=x['Name']), axis=1)
```
